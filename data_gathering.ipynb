```python
def to_lower(cell):
    if cell == "CALTECH":
        return "Caltech"
    elif cell == "LEUVEN_1":
        return "Leuven_1"
    elif cell == "LEUVEN_2":
        return "Leuven_2"
    elif cell == "MAX_MUN":
        return "MaxMun"
    elif cell == "OLIN":
        return "Olin"
    elif cell == "PITT":
        return "Pitt"
    elif cell == "STANFORD":
        return "Stanford"
    elif cell == "TRINITY":
        return "Trinity"
    elif cell == "YALE":
        return "Yale"
    else:
        return cell
    

df["SITE_ID"] = df["SITE_ID"].apply(to_lower) 
df[selected_columns]
```

```python
df_copy = df[selected_columns].copy()
df_copy.rename(columns={"SUB_ID": "Sub_id", "DX_GROUP": "Diagnosis", "AGE_AT_SCAN": "Age", "SEX": "Sex"}, inplace=True)
```

```python
def fmri_address(row):
    address = "../abide_data/Outputs/cpac/nofilt_noglobal/func_preproc/" + row["SITE_ID"] + "_00" + str(row["Sub_id"]) + "_func_preproc.nii.gz"
    return address


df_copy.loc[
    (df_copy["SITE_ID"] == "UM_2") & (df_copy["Sub_id"] == 50403),
    "Sub_id"
] = 50402
# df_copy = df_copy[~(((df_copy["SITE_ID"] == "Pitt") & (df_copy["Sub_id"] == 50004)) | ((df_copy["SITE_ID"] == "NYU") & (df_copy["Sub_id"] == 51040)))].reset_index(drop=True)
df_copy["fmri_file"] = df_copy.apply(fmri_address, axis=1)

print(df_copy.to_string())
```

```python

def tc_file_address(row):
    address = "../abide_data/Outputs/cpac/nofilt_noglobal/metadata/" + "00" + str(row["Sub_id"]) + "/tc/ATLAS/timecourse.csv"
    return address

def cc_file_address(row):
    address = "../abide_data/Outputs/cpac/nofilt_noglobal/metadata/" + "00" + str(row["Sub_id"]) + "/tc/ATLAS/corr_mat.npy"
    return address

def corrvec_file_address(row):
    address = "../abide_data/Outputs/cpac/nofilt_noglobal/metadata/" + "00" + str(row["Sub_id"]) + "/tc/ATLAS/corr_vec.npy"
    return address

df_copy["tc_file"] = df_copy.apply(tc_file_address, axis=1)
df_copy["cc_file"] = df_copy.apply(cc_file_address, axis=1)
df_copy["corrvec_file"] = df_copy.apply(corrvec_file_address, axis=1)
```
